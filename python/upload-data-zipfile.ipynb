```
import requests
import json
import pyspark.sql.functions as F
from pyspark.sql.types import StructField, StructType, StringType, ArrayType, LongType, TimestampType, BinaryType, IntegerType, DateType
from datetime import datetime
import time
import zipfile
import os
```

```
dbutils.widgets.text("fab", "fab") 
dbutils.widgets.text("period", "period") 
dbutils.widgets.text("source", "source") 

fab =  dbutils.widgets.get("fab")
period = dbutils.widgets.get("period")
source =  dbutils.widgets.get("source")

# fab = "D21"
# period = "D"
# source = "/mnt/source/D21_3S_TRANS_OIL_AZURE"
```

```
#pandas df
raw_df = spark.read.format('delta').load(source)

raw_df_pandas = raw_df.toPandas()
station_list = list(raw_df_pandas['STATION'].unique())
for name in station_list:
    print(name + ' ziptime :' + datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    G_list = list(raw_df_pandas[raw_df_pandas['STATION'] == name]['G'].unique())
    
    dbutils.fs.mkdirs(f'file:/databricks/driver/{name}')
    for g in G_list:
        select_name = raw_df_pandas['STATION'] == name
        select_g = raw_df_pandas['G'] == g
        raw_df_pandas[(select_name & select_g)][['timestamp','value']].sort_values(by=['timestamp']).reset_index(drop=True).to_csv(f'/databricks/driver/{name}/{g}.csv',index=False)  
          
    z = zipfile.ZipFile(f'/databricks/driver/{name}.zip', "w", zipfile.ZIP_DEFLATED)
    for f in os.listdir(f'/databricks/driver/{name}'):
        if f.endswith("csv"):
            z.write(f'/databricks/driver/{name}/{f}' , f )
    z.close()
    dbutils.fs.mv(f'file:/databricks/driver/{name}.zip', f'/mnt/zipfile/{fab}/{period}/{name}.zip')
    dbutils.fs.rm(f'file:/databricks/driver/{name}', True)
```
