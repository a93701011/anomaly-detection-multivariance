#### Access Azure Blob Storage with Databricks

Connect Azure Databricks cluster to data stored in an Azure storage account that has Azure Blob Storage enabled. This connection enables you to natively run queries and analytics from your cluster on your data.

```
blobStorage = "datalakecpcdev.blob.core.windows.net"
blobAccessKey =  dbutils.secrets.get(scope = 'cpc-keyvault-dev', key = 'datalakegne2-datalakecpcdev-key')
```

```
configs = {"fs.azure.account.key."+blobStorage : blobAccessKey}

blobContainer = 'source'
dbutils.fs.mount(
  source = "wasbs://" + blobContainer + "@"+ blobStorage,
  mount_point = "/mnt/source",
  extra_configs = configs)

blobContainer = 'deltalake'
dbutils.fs.mount(
  source = "wasbs://" + blobContainer + "@"+ blobStorage,
  mount_point = "/mnt/deltalake",
  extra_configs = configs)

blobContainer = 'zipfile'
dbutils.fs.mount(
  source = "wasbs://" + blobContainer + "@"+ blobStorage,
  mount_point = "/mnt/zipfile",
  extra_configs = configs)

configs = {"fs.azure.account.key."+blobStorage : blobAccessKey}
blobContainer = 'tempfile'
dbutils.fs.mount(
  source = "wasbs://" + blobContainer + "@"+ blobStorage,
  mount_point = "/mnt/tempfile",
  extra_configs = configs)
```

####Access Data Lake Storage Gen2 data with Azure Databricks

Connect Azure Databricks cluster to data stored in an Azure storage account that has Data Lake Storage Gen2 enabled. This connection enables you to natively run queries and analytics from your cluster on your data.

```
application_id = ""
authentication_id = ""
tenant_id = ""
```

```
storageAccountName = ""
fileSystemName = ""
```

```
configs = {"fs.azure.account.auth.type": "OAuth",
       "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
       "fs.azure.account.oauth2.client.id": application_id,
       "fs.azure.account.oauth2.client.secret": authentication_id,
       "fs.azure.account.oauth2.client.endpoint": "https://login.microsoftonline.com/{}/oauth2/token".format(tenant_id),
       "fs.azure.createRemoteFileSystemDuringInitialization": "true"}

dbutils.fs.mount(
source = "abfss://"+fileSystemName+"@"+storageAccountName+".dfs.core.windows.net",
mount_point = "/mnt/adls_gen2/",
extra_configs = configs)
```
